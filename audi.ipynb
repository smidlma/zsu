```python
fig = plt.figure(figsize=(16, 9))
sns.scatterplot(data = df, x = 'engineSize', y = 'mileage', hue = 'model')
```

Abych podpořil své předchozí tvrzení o tom, že nájezd aut nad 5 litrů by se neměl brát jako test odolnosti. 

Tak jsem si zobrazil, které modely mají motor s objemem nad 5 litrů. Vidíme tam modely jako RS4, RS5, R8 a RS7.

Všechny tyto modely jsou sportovním odvětvím Audi, které potřebují silné motory. Silný motor vede k vysoké spotřebě, proto tyto vozy nebudou tak moc ojeté jako vozy s dvoulitrovým motorem. 
Tudíž nemůžeme jejich nájezd, brát jako příklad kvality motoru.

```python
a = sns.displot(data = df, x = 'model', col = 'transmission')
a.set_xticklabels(rotation=75)
```

Zde se můžeme podívat na to, které modely a kolik z nich mělo různý druh převodovky. Zákazník chce například manuální převodovku. Kupuje-li starší vůz, tak se může podívat, jestli jeho model má dost zástupců s jeho variantou převodovky. Velikost sloupce odkazuje na cenu a sehnatelnost náhradních dílů. Čím více bylo aut s manuální převodovkou, tím menší má převodovka cenu. 

Naopak chce-li automatickou převodoku, tak se může podívat jestli vybraný model byl vyroben z velké části s automatickou převodovkou, což by naznačilo, že převodovka bude solidní. Jestli-že se model vyráběl s automatickou převodovkou pouze okrajově, může to odpovídat o tom, že Audi nedala do vývoje předovky v danném autě moc snahy a počítala spíše s manuální verzí pro tento model a dělala automat jen okrajově pro pár zákazníků. Jelikož automatická převodovka je náročnější a dražší na servis, je dobré si na tento fakt dávat pozor.

```python
fig = plt.figure(figsize=(16, 12))
sns.scatterplot(data = df, x = 'model', y = 'mpg', hue = 'fuelType')
```

Z tohoto grafu můžeme vyčíst jakou spotřebu mají jednotlivé modely. Všimněme si, že diesel má mnohem nižší spotřebu než benzín. Jsou zde i modely, které se prodávají například jen v dieselu a nebo pouze v benzínu.

.

```python
fig = plt.figure(figsize=(16, 12))
sns.scatterplot(data = df, x = 'year', y = 'mileage', hue = 'fuelType')
```

Z tohoto grafu můžeme vyčíst, že starší auta už vlastně nejde sehnat s nízkým nájezdem. Ve většině případech u všech ročníků aut má největší nájezd diesel. Může za to jeho nižší spotřeba a obliba jakožto auto pro denní ježdění. Je to také volba auta hlavně na delší trasy, což způsobí vysoký nájezd.

.

```python
sns.displot(df.fuelType, color = 'blue', edgecolor = 'black')
```

Zde vidíme, že nejvíce aut jezdí právě na diesel. Těsně za ním je benzín. Nejméně aut je zde hybrid. Může za to fakt, že je zde i spoustu starších aut z dob, kdy u danného modelu neexistovala varianta hybridu.
